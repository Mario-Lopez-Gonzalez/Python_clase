## EJERCICIO 2

````python
def toHex(RGB):
  tokens = RGB.split(",")
  myhex = "#"
  for token in tokens:
    myhex += format(int(token), '02x') # Formateamos el int([numero]) a un hexadecimal de dos cifras
  return myhex

print(toHex("15,255,14"))
````

````python
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
plt.figure(figsize=(8, 6))
fig, ax = plt.subplots()
node_size = [350 * G.degree(n) for n in G.nodes()] # Por cada nodo en nodos saca el grado y multiplicalo por 350 y metelo en una lista para node_size
nx.draw(G, ax=ax, node_color = "#ad80fa", edge_color = "#b1ddf5", font_color = "white", font_weight = "bold", with_labels = True, node_size = node_size)
plt.show()
````

## EJERCICIO 8

````python
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5), ("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
plt.figure(figsize=(8, 6))
fig, ax = plt.subplots()
node_size = [350 * G.degree(n) for n in G.nodes()] # Por cada nodo en nodos saca el grado y multiplicalo por 350 y metelo en una lista para node_size
edge_size = [2 * G[u][v]['weight'] for u, v in G.edges()] # Por cada conexion entre nodos (u,v) sacamos su weight con (u,v)["weight"] y lo multiplicamos por 2 y lo metemos en una lista para width
nx.draw(G, ax=ax, node_color = "#ad80fa", edge_color = "#b1ddf5", font_color = "white", font_weight = "bold", with_labels = True, node_size = node_size, width = edge_size)
plt.show()
````
